```python
import json
import pandas as pd
import random
```

```python
conv_path = '/data/LPJ/ICML25/GraphCoder/graphgpt_dataset/gpt_dataset_construction/stage2_1989_57/with_head/conversations.json'
graph_path = '/data/LPJ/ICML25/GraphCoder/graphgpt_dataset/gpt_dataset_construction/stage2_1989_57/with_head/graph.jsonl'
with open(conv_path, 'r') as f:
    conv = json.load(f)
graph = pd.read_json(graph_path, lines=True)
```

```python
# 假设 conv 和 graph 的长度相同，且索引一致
# 随机选取 20 个索引
selected_indices = random.sample(range(len(conv)), 20)

# 根据选取的索引提取数据
selected_conv = [conv[idx] for idx in selected_indices]
selected_graph = graph.iloc[selected_indices]

# 现在 selected_conv 和 selected_graph 包含了随机选取的 20 个相同 index 的数据
```
